```python
import pandas as pd
import numpy as np
import os
apple_stocks = pd.read_csv("apple_stock.csv")
apple_stocks
apple_stocks.head()
```

```python
apple_stocks.head()
```

```python
#removing unwanted columns
cleaned_duplicate_rows = apple_stocks.drop_duplicates()
cleaned_duplicate_rows.head()
```

```python
apple_stocks = apple_stocks.dropna()
apple_stocks.tail()
```

```python
apple_stocks = apple_stocks.replace(r'[\*\$\#]', '', regex=True) # 
apple_stocks
```

```python
apple_stocks = apple_stocks.rename(columns={'Unnamed: 0': 'Year'})
apple_stocks
```

```python
apple_stocks.groupby(['Year','Volume']).size()
```

```python
apple_stocks_sum = apple_stocks.groupby("Open")["Low"].sum().sort_values(ascending=False)
chart = apple_stocks_sum.head()
```

```python
chart.plot(kind="bar", title="Total Spent per Category", color='red')
```

```python
average_amount_spent_on_shopping  = apple_stocks['Volume'].mean()
print("Average Ammount Spent on Shopping is: ", average_amount_spent_on_shopping)
```

```python
chart.plot(kind="bar", title="Total Spent per Category", color='red')
```

```python
apple_stocks.groupby('High')['Volume'].sum().sort_values(ascending=False)
chart.head()
```

```python
chart.plot(kind="bar", title="Total Spent per Item in Shopping", figsize=(15, 10), color='skyblue')
```

```python
Volume_summary = apple_stocks.groupby(['Year', 'Volume']).sum()
Volume_summary
```

```python
chart.plot(kind="bar", title="Transaction Type in shopping", color='orange')
```

```python
chart.to_csv('Category_Summary.csv')
Volume_summary.to_csv('Volume_data.csv')
```
